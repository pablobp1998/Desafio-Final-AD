```python
import pandas as pd
import json
from pandas.io.json import json_normalize

with open('./Datos/datos_consejos.json') as data_file:    
    d= json.load(data_file)  
df= pd.DataFrame(d['data'])
df = df.drop(['index'], axis=1)
```

```python
years = []
for i in range(len(df['fecha'])):
    years.append(int(df['fecha'].iloc[i][0:4]))
df['fecha_corta'] = years
df = df.drop(['usuario_id','fecha', 'negocio_id', 'num_likes'], axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(df['fecha_corta'], bins= 8)
plt.show()
```

# Análisis de actividad

Si limitamos la fecha de los comentarios a los años, podremos observar la actividad que ha tenido la plataforma. Pese a que muchos usuarios llevan varios años utilizándola, podemos ver por el número de comentarios publicados en cada año que 2018 ha sido el año que más actividad de todos.

```python
df_spam = df[(df['usuario_id'] == "8uCfqWg3gs374IK-lBhRzw")].drop(['usuario_id','fecha', 'negocio_id', 'num_likes'], axis=1)
df_spam.groupby(df_spam.columns.tolist(),as_index=False).size().sort_values(ascending = False)
```

## Acceso a aplicación o farmeo

Observamos un comportamiento peculiar de un usuario. Utiliza la cuenta semi-regularmente desde 2014. Hasta mayo de 2018, ha escrito 111 reseñas aparentemente únicas. Sin embargo, el día 25 de mayo empieza a escribir el mismo mensaje, 30 veces en un período de 6 horas. 
Después de un mes sin actividad, escribe dos reseñas más, que parecen genuinas. Sin embargo, después de esto, tras casi un mes sin actividad, vuelve a escribir, esta vez en el intervalo de menos de un día, 61 reseñas idénticas.
Tras diez días sin escribir, escribirá dos reseñas cortas, relacionadas con el horario.
Tras 20 días de inactividad, escribe, de nuevo, 21 reseñas idénticas en menos de un día.
Por último, parece que vuelve a la normalidad, con reseñas aparentemente auténticas espaciadas en unos meses.

Este comportamiento parece indicativo de que el usuario ha dado acceso a alguna aplicación a su cuenta. Los comentarios escritos son únicos a este usuario dentro de los datos que tenemos y la gran mayoría de los comentarios se escriben a altas horas de la noche, por lo que es improbable que el usuario sea el que los está escribiendo. Sin embargo, los datos no parecen promocionar nada específico, por lo que no parece comportamiento propio de un _bot_.

Creemos que, o bien el usuario ha dado acceso a algún tipo de aplicación que escribe por él, o está _farmeando_ opiniones con alguna herramienta externa para conseguir recompensas de algún tipo.

# Usuarios más activos

También podemos obtener información del conjunto de datos a partir de los usuarios que más consejos tengan. El usuario más activo supera en casi 1000 opiniones a los demás, lo que supone una gran diferencia entre el primero y el segundo.

```python
df_users = pd.DataFrame(df['usuario_id'])
df_users.groupby(df_users.columns.tolist(),as_index=False).size().sort_values(ascending = False).head(50)
```

## Análisis de los tres mejores usuarios
Mientras que la diferencia entre el primero y el segundo es bastante notable, tras el segundo la diferencia es mucho más gradual, a veces de menos de cien comentarios. Hemos analizado los consejos que dan los tres primeros para ver si la información que aportan es de calidad.

Como se puede ver a continuación, los dos primeros usuarios repiten muy pocas veces sus opiniones. El usuario más activo da descripciones muy detalladas, e incluso algunas veces da una puntuación dentro de la propia opinión.
El segundo usuario más activo es más conciso. Da descripciones útiles sobre las características generales de los establecimientos: red de internet, precio medio, capacidad, etc. Por lo tanto, hay varios comentarios que están repetidos o son ligeras variaciones de una misma opinión.
El tercer usuario con más consejos es el que menos información aporta. Tiene una gran cantidad de reseñas con menos de tres palabras y muchas están repetidas varias veces, probablemente por lo generales que son.
